```python
import warnings
warnings.filterwarnings('ignore')
import transformers
transformers.logging.set_verbosity_error()
```

Load the model for Translation.

```python
from transformers import MarianMTModel, MarianTokenizer
```

```python
model_name = 'Helsinki-NLP/opus-mt-tr-en'
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)
```

```python
sentence = """İspanya ve Portekiz, son bir ay içinde ikinci kez sıcak hava dalgasının etkisine girerken Fransa ve 
İngiltere'de de önümüzdeki günlerde kavurucu sıcaklıkların yaşanması bekleniyor."""
```

```python
print(tokenizer([sentence]).input_ids)
```

```python
input_ids = tokenizer([sentence], return_tensors="pt").input_ids
outputs = model.generate(input_ids=input_ids, num_beams=5, num_return_sequences=1)
print("Generated:", tokenizer.batch_decode(outputs, skip_special_tokens=True))
```
